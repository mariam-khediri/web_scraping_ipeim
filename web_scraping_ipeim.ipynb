```python
from bs4 import BeautifulSoup
import requests
```

```python
url ="https://ipeim.rnu.tn/fra/enseignants"
```

```python
response = requests.get(url)
```

```python
data = response.text
```

```python
soup = BeautifulSoup(data,'html.parser')
```

```python
profiles = soup.find_all('div',{'class':'tcard-front'})
```

```python
l = []
for profile in profiles : 
    #name
    name = profile.find('div',{'class':'h2'}).text.strip()
    print(name)
    #mail
    mail = profile.find_all('a')[1]
    adress = 'None'
    if len(mail.get('href').strip()) > 7  : 
        #empty strings are of ((length == 7)) in my uni's website
        #uncomment the next line to know more
        
        #print(len(mail.get('href').strip()))
        adress = mail.get('href')[7:]
    print(adress)
    #web 
    web_page = profile.find_all('a')[0].get('href')
    print(web_page)
    
    l.append([name,adress,web_page]) 
    
    print('-------------')
```

```python
import pandas as pd 
df = pd.DataFrame(l,columns=['name','email','webpage'])
```
